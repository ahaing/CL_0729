```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
student1 = {'CH':87, 'EN':90, 'MA':100, 'SC':200}
student2 = {'CH':100, 'EN':108, 'MA':100,'SC':108 }
student1.keys()
subjects = list(student1.keys())
subjects

value1=list(student1.values())
value2=list(student2.values())
```

```python
figure = plt.figure(figsize=(8.0,5.0))
ax = figure.add_subplot()
ax.plot(subjects, value1)
ax.plot(subjects, value2, 'r',linestyle='dashed') # set_linestyle(ls) # ':' or 'dotted'=>' r:'
ax.set_title("student 1 & 2 comparison")
plt.show() #google matplotlib.axes.subplot(?)
# matplotlib.axes.Axes.plot
```
